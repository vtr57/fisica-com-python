# Movimento em 2 dimensões

Trataremos nas aulas2.x do lançamento horizontal e do lançamento oblíquo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,1,0.05)
x_i, y_i, v_i, g = 0, 10, 2, 10
x = v_i*t + x_i
y = -g*t**2
```

```python
fig, ax = plt.subplots()
ax.plot(t, x)
ax.set_xlabel('tempo (s)')
ax.set_ylabel('posição horizontal (m)')
```

```python
fig2, ax2 = plt.subplots()
ax2.plot(t, y)
ax2.set_xlabel('tempo (s)')
ax2.set_ylabel('posição vertical (m)')
```

```python
fig3, ax3 = plt.subplots()
ax3.plot(x, y)
ax3.set_xlabel('posição horizontal (m)')
ax3.set_ylabel('posição vertical (m)')
```
