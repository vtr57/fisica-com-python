# Movimento Circular Uniforme

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r, ohmega, T = 5, 2, 2*np.pi/ohmega
t = np.arange(0,T,0.1)
x = r*np.cos(ohmega*t)
y = r*np.sin(ohmega*t)
```

```python
fig, ax = plt.subplots()
ax.scatter(t, x, c='r', label = 'dados', s=10)
ax.set_xlabel('tempo (s)')
ax.set_ylabel('posição horizontal (m)')
```

```python
fig, ax = plt.subplots()
ax.scatter(t, y, c='r', label = 'dados', s=10)
ax.set_xlabel('tempo (s)')
ax.set_ylabel('posição vertical (m)')
```

```python
fig, ax = plt.subplots()
ax.scatter(x, y, c='r', label = 'dados', s=2)
ax.set_xlabel('posição horizontal (m)')
ax.set_ylabel('posição vertical (m)')
fig.set_size_inches(5,5)
```
