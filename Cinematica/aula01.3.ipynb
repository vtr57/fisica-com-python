# Gráfico do problema da ultrapassagem

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xai, va, xbi, vb, t_max, n = 20, 10, 100, 2, 20, 1000
tempo = np.linspace(0, t_max, n)
xa = va*tempo + xai
xb = vb*tempo + xbi
```

```python
fig, ax = plt.subplots()
ax.plot(tempo, xa)
ax.plot(tempo, xb)
ax.set_title('Ultrapassagem')
ax.set_ylabel('Posição')
ax.set_xlabel('Tempo')
ax.legend(['Partícula A', 'Partícula B'])
```

Muitas vezes, para resolver algum problema, você ganhará mais intimidade realizando um gráfico primeiro antes de querer descobrir algum valor.
