# Movimento em 1-D

O movimento em 1 dimensão é o movimento mais simples que se pode trabalhar. Com ele conseguiremos aprender alguns princípios da Mecânica de maneira bem simples.

## Gerando os Dados

```python
x = 10  # x = 10m
dx = 2  # dx = 2m
t_max = 10  #t_max = 10s
```

```python
tempo = [] 
for i in range(t_max + 1): 
  tempo.append(i)
```

```python
posicao = [] #lista que guardará os valores de posição
for i in range(t_max + 1): #executa um looping (t_max + 1) vezes
  posicao.append(x) #adiciona x na lista da posição
  x += dx #incremento da posição x com o incremento dx
```

```python
posicao
```

```python
tempo
```

## Visualizando os dados em um gráfico

Para visualizar os dados em um gráficos utilizaremos o framework chamado Matplotlib. Para isso precisamos importar esse framework com o seguinte código.

```python
import matplotlib.pyplot as plt
# 'import matplotlib.pyplot' chamar as classes e objetos que estão no arquivo matplotlib da classe pyplot
# 'as plt' é o nome que usaremos em código da classe matplotlib.pyplot
```

```python
fig, ax = plt.subplots()
ax.plot(tempo, posicao)
ax.set_title('Movimento em 1-D')
ax.set_ylabel('Posição')
ax.set_xlabel('Tempo')
ax.set_ylim(0, 35)
ax.set_xlim(0, 35)
```

```python
fig2, ax2 = plt.subplots()
fig2.set(figheight=3, figwidth=3)
ax2.scatter(tempo, posicao)
ax2.set_title('Movimento em 1-D')
ax2.set_ylabel('Posição')
ax2.set_xlabel('Tempo')
ax2.set_ylim(0, 35)
ax2.set_xlim(0, 35)
```
