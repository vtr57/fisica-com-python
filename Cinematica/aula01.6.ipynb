# Movimento Acelerado

Vamos estudar os gráficos do movimento acelerado e constante, os diferentes sinais e suas consequências para os gráficos.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0, 10, 0.1)
a, vi, xi = 2, 0, 3
v = a*t + vi
x = a/2*t**2 + vi*t + xi
```

```python
fig, ax = plt.subplots()
ax.plot(t, v)
ax.set_xlabel('tempo (s)')
ax.set_ylabel('velocidade (m/s)')
```

```python
fig, ax = plt.subplots()
ax.plot(t, x)
ax.set_xlabel('tempo (s)')
ax.set_ylabel('posição (m)')
```
