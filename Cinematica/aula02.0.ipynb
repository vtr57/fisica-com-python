# Movimento em 2 dimensões

Trataremos nas aulas2.x do lançamento horizontal e do lançamento oblíquo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(0,1,0.05)
x_i, y_i, v_i, g = 0, 10, 2, 10
x = v_i*t + x_i
y = -g*t**2 + y_i
```

```python
fig, ax = plt.subplots()
ax.scatter(t, x, marker='o', c='r', label = 'dados')
ax.set_xlabel('tempo (s)')
ax.set_ylabel('posição horizontal (m)')
```

```python
fig2, ax2 = plt.subplots()
ax2.scatter(t, y, marker='o', c='r', label = 'dados')
ax2.set_xlabel('tempo (s)')
ax2.set_ylabel('posição vertical (m)')
```

```python
fig3, ax3 = plt.subplots()
ax3.scatter(x, y, marker='o', c='r', label = 'dados')
ax3.set_xlabel('posição horizontal (m)')
ax3.set_ylabel('posição vertical (m)')
```

```python
# econtrando a equação da curva do gráfico yXx
p = np.polyfit(x, y, 2)
fit3 = np.polyval(p, x)
```

```python
fig4, ax4 = plt.subplots()
ax4.scatter(x, y, marker='o', c='r', label = 'dados')
ax4.plot(x, fit3, label='fit')
ax4.set_xlabel('posição horizontal (m)')
ax4.set_ylabel('posição vertical (m)')
ax4.legend()
```
